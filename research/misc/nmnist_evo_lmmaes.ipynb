# Neuroevolution of an SNN on N-MNIST (LM-MA-ES)

This notebook trains a spiking neural network on **Neuromorphic-MNIST** (N-MNIST),
the event-camera version of MNIST, **without gradients**. The network weights are
optimized with the **LM-MA-ES** evolution strategy instead of backpropagation
through a surrogate spike function.

It is modernized from the original Haiku + old-`evosax` implementation to
**Flax NNX + evosax 0.2**:

* a feed-forward SNN (`spyx.nn.Sequential` of `nnx.Linear` + spiking neurons),
  run over time with `spyx.nn.run`;
* parameters flattened to a single vector via
  `nnx.split` + `jax.flatten_util.ravel_pytree` for the ES to search;
* evolution via `evosax.algorithms.distribution_based.LM_MA_ES`.

As in the original, a simple **dense** architecture is used (the event frames are
flattened) rather than a convolutional one, so that a whole population of networks
can be evaluated cheaply. This is a deliberately **CPU-tiny** run: small population,
few generations, a handful of prestaged batches. The goal is to show the pipeline
runs and that the elite fitness **improves generation over generation**, not to
reach a competitive accuracy.

```python
import os
import time

os.environ.setdefault("JAX_PLATFORMS", "cpu")  # CPU-only run

import jax
import jax.numpy as jnp
import numpy as np
from evosax.algorithms.distribution_based import LM_MA_ES
from flax import nnx

import spyx
import spyx.nn as snn

# N-MNIST event frames are (2, 34, 34); flattened that is the dense input dim.
SAMPLE_T, IN_DIM, N_CLASSES, HIDDEN = 15, 2 * 34 * 34, 10, 64
BATCH = 32
POP, GENS = 24, 15  # tiny scale for a CPU sanity run
N_FIT_BATCHES = 4  # prestaged batches used to score each candidate
N_EVAL_BATCHES = 4  # prestaged held-out batches for elite evaluation

print(f"backend={jax.default_backend()}  pop={POP}  gens={GENS}")
```

## Data

N-MNIST is loaded with `spyx.data.NMNIST_loader`, which yields spike frames
bit-packed along the time axis (`.obs`) together with integer `.labels`. We
**prestage a small number of batches once** into dense float arrays and reuse them
for every generation (rather than re-reading the dataset each generation). Each
frame is unpacked along the time axis and flattened from `(2, 34, 34)` to a
`2*34*34` input vector.

The loader's *test* split is stored ordered by class (every early batch is a
single digit), which would make a tiny held-out set meaningless. So we instead
draw both the fitness batches and a separate block of **held-out evaluation
batches** from the *shuffled train* stream: the fitness batches drive selection,
the held-out (mixed-class) batches only measure the elite's generalization.

```python
def stage(loader_iter, n_batches):
    """Pull n_batches, unpack the time axis, flatten frames to (B, T, IN_DIM)."""
    obs_list, lbl_list = [], []
    for _ in range(n_batches):
        b = next(loader_iter)
        up = np.unpackbits(np.asarray(b.obs), axis=1)[:, :SAMPLE_T]  # (B,T,2,34,34)
        up = up.reshape(up.shape[0], SAMPLE_T, IN_DIM).astype(np.float32)
        obs_list.append(up)
        lbl_list.append(np.asarray(b.labels))
    return jnp.asarray(np.stack(obs_list)), jnp.asarray(np.stack(lbl_list))


def find_data_dir():
    """Locate the already-downloaded ./data/NMNIST by walking up from the cwd.

    nbconvert runs the kernel with the notebook's directory as cwd, so a bare
    "./data" would point at research/misc/data and trigger a 2.6 GB re-download.
    Walking up finds the repo-root data dir whether run as a script or a notebook.
    """
    d = os.getcwd()
    for _ in range(6):
        if os.path.isdir(os.path.join(d, "data", "NMNIST")):
            return os.path.join(d, "data")
        d = os.path.dirname(d)
    return "./data"  # fall back to default (will download if absent)


DATA_DIR = find_data_dir()

# worker_count=0 keeps everything in-process (safe under nbconvert; data is cached).
t0 = time.perf_counter()
dl = spyx.data.NMNIST_loader(
    batch_size=BATCH, sample_T=SAMPLE_T, download_dir=DATA_DIR, worker_count=0
)
train_it = dl.train_epoch()  # shuffled stream
fit_obs, fit_lbl = stage(train_it, N_FIT_BATCHES)  # for selection
eval_obs, eval_lbl = stage(train_it, N_EVAL_BATCHES)  # held-out, for scoring the elite
print(
    f"prestaged fit_obs={fit_obs.shape} eval_obs={eval_obs.shape} "
    f"in {time.perf_counter() - t0:.1f}s"
)
```

## Model

A dense feed-forward SNN: a `LIF` hidden layer and a leaky-integrator (`LI`)
readout, matching the original architecture (shrunk to a 64-unit hidden layer for
CPU). We build one instance to get the parameter tree, then flatten it into the
vector the ES optimizes.

```python
def make_model(rngs):
    return snn.Sequential(
        nnx.Linear(IN_DIM, HIDDEN, use_bias=False, rngs=rngs),
        snn.LIF((HIDDEN,), rngs=rngs),
        nnx.Linear(HIDDEN, N_CLASSES, use_bias=False, rngs=rngs),
        snn.LI((N_CLASSES,), rngs=rngs),
    )


graphdef, params = nnx.split(make_model(nnx.Rngs(0)), nnx.Param)
flat, unravel = jax.flatten_util.ravel_pytree(params)
print(f"evolving {flat.size} parameters")
```

## Fitness function

Each candidate is a flat weight vector: rebuild the model with `nnx.merge`, run it
over time with `spyx.nn.run` (time-major `(T, B, C)`), and score with
`spyx.fn.integral_accuracy` (higher = better). `pop_fitness` vmaps this over the
whole population.

```python
acc_fn = spyx.fn.integral_accuracy()


def run_model(flat_params, x_BTC):
    model = nnx.merge(graphdef, unravel(flat_params))
    traces, _ = snn.run(model, jnp.transpose(x_BTC, (1, 0, 2)))  # (T,B,C) in
    return jnp.transpose(traces, (1, 0, 2))  # -> (B,T,C)


def fitness(flat_params, obs, lbl):  # higher = better (mean accuracy)
    accs = jnp.stack(
        [acc_fn(run_model(flat_params, obs[i]), lbl[i])[0] for i in range(obs.shape[0])]
    )
    return jnp.mean(accs)


pop_fitness = jax.jit(jax.vmap(fitness, in_axes=(0, None, None)))
eval_elite = jax.jit(fitness)
```

## Evolution loop (LM-MA-ES, evosax 0.2)

Each generation: `ask` for candidate weight vectors, evaluate their accuracy, then
`tell` the strategy the **negated** accuracy (evosax *minimizes*). We track the
elite's held-out evaluation accuracy every few generations.

```python
strat = LM_MA_ES(population_size=POP, solution=jnp.zeros_like(flat))
es_p = strat.default_params
es_s = strat.init(jax.random.PRNGKey(0), mean=flat, params=es_p)

key = jax.random.PRNGKey(1)
t0 = time.perf_counter()
history = []
for g in range(1, GENS + 1):
    key, k_ask, k_tell = jax.random.split(key, 3)
    cands, es_s = strat.ask(k_ask, es_s, es_p)
    fit = pop_fitness(cands, fit_obs, fit_lbl)  # accuracy per candidate
    es_s, _ = strat.tell(k_tell, cands, -fit, es_s, es_p)  # negate: evosax minimizes
    if g % 3 == 0 or g == 1:
        eval_acc = float(eval_elite(es_s.best_solution, eval_obs, eval_lbl))
        history.append((g, float(jnp.max(fit)), eval_acc))
        print(
            f"gen {g:3d}  best_pop_fit_acc={float(jnp.max(fit)) * 100:5.2f}%  "
            f"elite_eval_acc={eval_acc * 100:5.2f}%  ({time.perf_counter() - t0:.0f}s)"
        )
```

## Results

The elite evaluation accuracy should climb from its generation-1 value. N-MNIST
has 10 classes (chance 10%); at this tiny CPU scale the evolved network only edges
above chance, which is expected — the point is that LM-MA-ES **improves** the SNN
over generations.

```python
final = float(eval_elite(es_s.best_solution, eval_obs, eval_lbl))
print(f"FINAL evolved SNN: elite_eval_acc={final * 100:.2f}%  ({GENS} gens)")
print("gen / best_pop_fit_acc / elite_eval_acc history:")
for g, tr, te in history:
    print(f"  gen {g:3d}:  fit {tr * 100:5.2f}%   eval {te * 100:5.2f}%")
```
